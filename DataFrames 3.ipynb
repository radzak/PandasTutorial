```python
bond = pd.read_csv('datasets/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
bond.head(3)
```

```python
bond.ix['GoldenEye']  # same output as bond.loc['GoldenEye']
```

# A Review of the `.apply()` Method on Single Columns

```python
bond = pd.read_csv('datasets/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
```

### manual way, one by one

```python
def convert_to_string_and_add_millions(number):
    return str(number) + ' MILLIONS!'

bond['Box Office'] = bond['Box Office'].apply(convert_to_string_and_add_millions)
```

```python
bond['Budget'] = bond['Budget'].apply(convert_to_string_and_add_millions)
```

### faster way, applying function to multiple columns simultaneously (for loop)

```python
bond = pd.read_csv('datasets/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
```

```python
columns = ['Box Office', 'Budget', 'Bond Actor Salary']
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head()
```

# The `.apply()` Method with Row Values

```python
bond = pd.read_csv('datasets/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
```

### applying function to each row

```python
def good_movie(row):
    actor = row['Actor']  # actor = row[1]
    budget = row['Budget']  # budget = row[4]
    
    if actor == 'Pierce Brosnan':
        return 'The best'
    elif actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    else:
        return 'I have no clue'
    
bond.apply(good_movie, axis='columns')
# this is confusing!! think of it as moving across the columns (horizontally) within a single row
# 'collecting' values from each columnd and creating a Series:
# pd.Series({'Year': 1985, 'Actor': 'Roger Moore', ...})
# you can access it both with numeric indexes and keys inside of a function that you want to apply to all rows
```

### applying function to each column

```python
def change_column(col):
    # col is a Series
    col = col.astype('str')
    col += ' haHAA'
    return col

bond.apply(change_column)
# axis = 0 or 'index': apply function to each column
```

# The `.copy()` Method

```python
bond = pd.read_csv('datasets/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
```

```python
directors = bond['Director']
directors.head(3)
```

```python
bond = pd.read_csv('datasets/jamesbond.csv', index_col='Film')
bond.sort_index(inplace=True)
```

```python
directors = bond['Director'].copy()  # separated, unique Series object
directors.head(3)
```

```python
directors['A View to a Kill'] = 'Mister John Glen'
directors.head(3)
```
